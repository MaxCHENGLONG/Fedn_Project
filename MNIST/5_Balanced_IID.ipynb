```python
import json
import matplotlib.pyplot as plt

# 
file_paths = [
    "/Users/max/Fedn_Project/MNIST/5clients_balanced_Alpha=10/metadata_1_10.0.json",
    "/Users/max/Fedn_Project/MNIST/5clients_balanced_Alpha=10/metadata_2_10.0.json",
    "/Users/max/Fedn_Project/MNIST/5clients_balanced_Alpha=10/metadata_3_10.0.json",
    "/Users/max/Fedn_Project/MNIST/5clients_balanced_Alpha=10/metadata_4_10.0.json",
    "/Users/max/Fedn_Project/MNIST/5clients_balanced_Alpha=10/metadata_5_10.0.json"
]

#  local_path_norm  global_path_norm
all_local_path_norms = []
global_path_norms = []

#  local_path_norm  global_path_norm 
for idx, file_path in enumerate(file_paths):
    with open(file_path, 'r') as file:
        data = json.load(file)
        # 
        if isinstance(data, list):
            #  local_path_norm 
            local_path_norms = [entry['local_path_norm'] for entry in data]
            # 
            local_path_norms.insert(0, 5.903397471741662)
            #  global_path_norm
            if idx == 0:
                global_path_norms = [entry['global_path_norm'] for entry in data]
        else:
            # 
            local_path_norms = [data]
        all_local_path_norms.append(local_path_norms)

# 
colors = ['blue', 'green', 'red', 'orange', 'purple']
linestyles = ['-', '--', '-.', ':', '-']
markers = ['o', 's', '^', 'D', 'v']

#  local_path_norm 
for i, local_path_norms in enumerate(all_local_path_norms):
    iterations = range(len(local_path_norms))
    plt.plot(iterations, local_path_norms,
             linestyle=linestyles[i],
             color=colors[i],
             marker=markers[i],
             label=f'Client {i+1}')

#  global_path_norm 
iterations = range(len(global_path_norms))
plt.plot(iterations, global_path_norms,
         linestyle='-', color='black', marker='*', label='Global')

plt.xlabel('Iteration')
plt.ylabel('Path Norm')
plt.title('Local and Global Path Norms (5 Clients, 50 Rounds, Balanced IID)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt

def plot_label_distribution(labels, title='Label Distribution'):
    """
    

    
    labels (torch.Tensor  np.array): 
    title (str): 
    """
    #  NumPy 
    labels_np = labels.numpy() if isinstance(labels, torch.Tensor) else np.array(labels)

    #  NumPy 
    unique_labels, counts = np.unique(labels_np, return_counts=True)

    # 
    total_count = len(labels_np)
    for label, count in zip(unique_labels, counts):
        percentage = (count / total_count) * 100
        print(f"Label: {label}, Count: {count}, Percentage: {percentage:.2f}%")

    # 
    plt.bar(unique_labels, counts)
    plt.xlabel('Labels')
    plt.ylabel('Count')
    plt.title(title)
    plt.xticks(unique_labels)
    plt.show()

data = torch.load('/Users/max/Fedn_Project/5clients_balanced_Alpha=10/clients_data/1/mnist.pt')
y_train = data['y_train']
plot_label_distribution(y_train, title='Label Distribution in Training Set (y_train)')
```
