```python
import matplotlib.pyplot as plt
fig, ax1 = plt.subplots(figsize=(50, 6))  # Width of 50, height of 6
epochs0 = range(1, len(test_accuracy0) + 1)
epochs1 = range(1, len(test_accuracy1) + 1)

# Plot accuracy on the left y-axis
ax1.plot(epochs0, train_accuracy0, label='Train Accuracy 0', color='blue', marker='o')
ax1.plot(epochs1, train_accuracy1, label='Train Accuracy 1', color='cyan', marker='x')
ax1.plot(epochs0, test_accuracy0, label='Test Accuracy 0', color='green', marker='s')
ax1.plot(epochs1, test_accuracy1, label='Test Accuracy 1', color='lime', marker='d')
ax1.set_xlabel("Epoch")
ax1.set_ylabel("Accuracy", color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Set x-axis ticks at every step
ax1.set_xticks(range(1, len(path_norms0) + 1))

# Add vertical lines every 3 steps to mark significant epochs
for x in range(0, len(path_norms0), 3):
    if x == 0:
        continue
    ax1.axvline(x=x, color='red', linestyle='--', linewidth=1.0)

# Plot loss on the right y-axis
ax2 = ax1.twinx()  # Create a second y-axis that shares the same x-axis
ax2.plot(epochs0, train_loss0, label='Train Loss 0', color='orange', marker='^')
ax2.plot(epochs1, train_loss1, label='Train Loss 1', color='red', marker='v')
ax2.plot(epochs0, test_loss0, label='Test Loss 0', color='magenta', marker='p')
ax2.plot(epochs1, test_loss1, label='Test Loss 1', color='purple', marker='h')
ax2.set_ylabel("Loss", color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Add legends for both y-axes
fig.legend(loc="upper right", bbox_to_anchor=(1, 1), bbox_transform=ax1.transAxes)

# Chart title and grid
plt.title("Model Metrics Comparison Non-IID Balanced Resnet FedOpt 2 Clients")
plt.grid()

# Show the chart
plt.show()
```
